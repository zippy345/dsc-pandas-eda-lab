### Sale Price

In the cell below, produce a histogram for `SalePrice`.

````python
# write a code that plots a histogram for the distribution of the SalePrice, TotRmsAbvGrd, and OverallCond.
# Define a function to plot histograms with mean line
def plot_histogram(data, column, title, x_label, y_label):
    data[column].plot(kind="hist", bins= 60, figsize =(10,6), color="green", edgecolor='black', alpha= 0.8)
    mean_value= data[column].mean()
    plt.axvline(mean_value, color='orange', linestyle='solid', linewidth=2, label=f'Mean: {mean_value:.2f}')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    plt.show()
````

## 3. Explore Differences between Subsets

As you might have noted in the previous step, the overall condition of the house seems like we should treat it as more of a categorical variable, rather than a numeric variable.

One useful way to explore a categorical variable is to create subsets of the full dataset based on that categorical variable, then plot their distributions based on some other variable. Since this dataset is traditionally used for predicting the sale price of a house, let's use `SalePrice` as that other variable.

In the cell below, create three variables, each of which represents a record-wise subset of `df` (meaning, it has the same columns as `df`, but only some of the rows).

* `below_average_condition`: home sales where the overall condition was less than 5
* `average_condition`: home sales where the overall condition was exactly 5
* `above_average_condition`: home sales where the overall condition was greater than 5

````python
#write a code that calculates the below,average and above average conditions of the OverallCond column
below_average_condition =df[df['OverallCond']<5]
average_condition = df[df['OverallCond']==5]
above_average_condition = df[df['OverallCond']>5]
````

The following code checks that you created the subsets correctly:

````python
# Run this cell without changes

# Check that all of them still have 80 columns
assert below_average_condition.shape[1] == 80
assert average_condition.shape[1] == 80
assert above_average_condition.shape[1] == 80

# Check the numbers of rows of each subset
assert below_average_condition.shape[0] == 88
assert average_condition.shape[0] == 821
assert above_average_condition.shape[0] == 551
````

The following code will produce a plot of the distributions of sale price for each of these subsets:

````python
# Run this cell without changes

# Set up plot
fig, ax = plt.subplots(figsize=(15,5))

# Create custom bins so all are on the same scale
bins = range(df["SalePrice"].min(), df["SalePrice"].max(), int(df["SalePrice"].median()) // 20)

# Plot three histograms, with reduced opacity (alpha) so we
# can see them overlapping
ax.hist(
    x=above_average_condition["SalePrice"],
    label="above average condition",
    bins=bins,
    color="blue",
    alpha=0.5
)
ax.hist(
    x=average_condition["SalePrice"],
    label="average condition",
    bins=bins,
    color="green",
    alpha=0.3
)
ax.hist(
    x=below_average_condition["SalePrice"],
    label="below average condition",
    bins=bins,
    color="yellow",
    alpha=0.5
)

# Customize labels
ax.set_title("Distributions of Sale Price Grouped by Condition")
ax.set_xlabel("Sale Price")
ax.set_ylabel("Number of houses")
ax.legend()
plt.show()
````

#### For deeper understanging i will conduct a groupby function on the `OverallCond` and the `SalePrice` columns the find the mean of saleprice for each Overallcondition.this will enable me to have a vivid picture of the conditions and the saleprice of the houses in Ames Housing dataset.   

````python
# Write a code using the Groupby function to group the OverallCond by SalePrice
Mean_price_tag= df.groupby(df['OverallCond'])['SalePrice'].mean()
Mean_price_tag
````

````python
#print out a barchart using the data above and the matplotlib library.
plt.figure(figsize=(10,6))
plt.bar(Mean_price_tag.index.astype(str), Mean_price_tag.values, color='green', edgecolor='black')
plt.title("Average SalePrice by OverallCond")
plt.xlabel("Overall Condition" )
plt.ylabel("Average SalePrice")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

````

````python

# Get only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation of all numeric columns with 'SalePrice'
correlations_with_saleprice = numeric_df.corr()['SalePrice'].drop('SalePrice')

# Find the column with the highest positive correlation
max_corr_column = correlations_with_saleprice.idxmax()
max_corr_value = correlations_with_saleprice.max()

print(f"The column most positively correlated with SalePrice is '{max_corr_column}' with a Pearson correlation of {max_corr_value:.4f}")
````

Now, find the ***most negatively correlated*** column:

````python
# Write a code that prints out the  most negatively correlated 
# Get only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation of all numeric columns with 'SalePrice'
correlations_with_saleprice = numeric_df.corr()['SalePrice'].drop('SalePrice')

# Find the column with most negative correlation
min_corr_column = correlations_with_saleprice.idxmin()
min_corr_value = correlations_with_saleprice.min()

print(f"The column most negatively correlated with SalePrice is '{min_corr_column}' with a Pearson correlation of {min_corr_value:.4f}")
````
